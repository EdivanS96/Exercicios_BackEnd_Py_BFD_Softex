```python
'''
Lista de Revisão do Curso Back-End com Python do progama BFD Softex
Link:https://github.com/Jose-George/python-bolsa-futuro-digital/blob/9496a6fb5b8e4ec8ddc843a940c51f0f6782a805/aula%20-%2006/Lista%20de%20Revis%C3%A3o%20Geral.md
aula - 06/Lista de Revisão Geral.md
'''

# 1 - Lista
lista = [1, 2, 3, 4, 5]
for elemento in lista:
    print(elemento)

# 2 - Condicinonais
numero = int(input("Digite um número: "))
if numero > 0:
    print("Positivo")
elif numero < 0:
    print("Negativo")
else:
    print("Zero")

#3 - Laços
for i in range(1, 11):
    print(i)

#4 - dicionatio
pessoas = {
    "Pessoa1": 25,
    "Pessoa2": 30,
    "Pessoa3": 22
}
for nome, idade in pessoas.items():
    print(f"{nome}: {idade} anos")
#5 Funções
def soma(a, b):
    return a + b
resultado = soma(5, 3)
print(resultado)

#6 - Strings
palavra = input("Digite uma palavra: ")
print(f"A palavra '{palavra}' tem {len(palavra)} caracteres.")

#7 Listas com 3 frutas
frutas = ["maçã", "banana", "laranja"]
frutas.append("uva")
print(frutas)

#8 condicionais
numero = int(input("Digite um número: "))
if numero % 2 == 0:
    print("Par")
else:
    print("Ímpar")

# 9 Laços com loop:

contador = 1
while contador <= 5:
    print(contador)
    contador += 1

#10 Dicionario:
produtos = {
    "Camiseta": 20,
    "Calça": 30,
    "Tênis": 50
}
for produto, preco in produtos.items():
    print(f"{produto}: R${preco}")

#11 Lista + Laços:
numeros = [1, 2, 3, 4, 5]
soma = 0
for numero in numeros:
    soma += numero
print(f"A soma dos números é: {soma}")

#12 Dicionarios com Tuplas:
pessoas = {
    "João": (25, "São Paulo"),
    "Maria": (30, "Rio de Janeiro"),
    "Pedro": (22, "Belo Horizonte")
}

#13 Condicionais Alinhadas (Pequeno, Médio ou Grande):
numero = int(input("Digite um número: "))
if numero < 10:
    print("Pequeno")
elif numero < 50:
    print("Médio")
else:
    print("Grande")

#14 Laços aninhados (Tabuada de 1 a 5):
for i in range(1, 6):
    for j in range(1, 6):
        print(f"{i} x {j} = {i * j}")

#15 Funções + Listas: recebe uma lista e retorna o maior numero:
def encontrar_maior(lista):
    maior = lista[0]
    for numero in lista:
        if numero > maior:
            maior = numero
    return maior

#16 Dicionario: conta a frequencia de letra na palavra: Hello ex: H:1, e:1, l:2, o:1
def contar_letras(palavra):
    frequencia = {}
    for letra in palavra:
        if letra in frequencia:
            frequencia[letra] += 1
        else:
            frequencia[letra] = 1
    return frequencia
```
